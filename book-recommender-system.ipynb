```python
# Reading the dataset
ratings = pd.read_csv('dataset/Ratings.csv', error_bad_lines=False, encoding='latin-1')
```

```python
ratings.rename(columns={'User-ID':'user_id', 'ISBN':'ISBN', 'Book-Rating':'rating'}, inplace=True)
```

```python
# Preprocessing 
# Consider Users who have rated atleast 200 books
# Consider Books which have been given rating by atleast 50 users
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
x[x] # taking only true values
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
books.head()
```

```python
# Creating new dataframe. with books and ratings
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'}, inplace=True)
```

```python
number_rating.head()
```

```python
# merge with ratings_with_books
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.rename(columns={'number of ratings':'number_of_ratings'}, inplace=True)
```

```python
final_rating.head(2)
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.shape
```

```python
# SO there are "61853" ratings done by those "Users" who have rated more than 200 books
# and on those books which have more then 50 ratings
```

```python
# droping duplicate values
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
# So There are 742 books rated by 888 Users
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csc_matrix
```

```python
# Will create books_sparse matrix because book_pivot lot of "0" values 
# If we give book_pivot dataset to NearestNeighbors algorithm then it will calculate distance of "0" values also 
books_sparse = csc_matrix(book_pivot)
```

```python
books_sparse
```

```python
# Model building
from sklearn.neighbors import NearestNeighbors
```

```python
# algorithm='brute' : calculate for all
model = NearestNeighbors(algorithm='brute') 
```

```python
model.fit(books_sparse)
```

```python
book_pivot.iloc[237, :].values # Harry poter book
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
book_pivot.index[237]
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    for i in range(len(suggestions)):
        if i == 0:
            print("The suggestion for", book_name, "are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
        
```

```python
recommend_book('Animal Farm')
```
